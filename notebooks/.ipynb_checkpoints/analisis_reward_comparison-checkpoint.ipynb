# Análisis de comparación de funciones de recompensa (DQN trading agent)

**Stage:** `reward_comparison` · **Intervalo:** 1h · **Configuración única de hiperparámetros:** `lambda_tc=0.02`, `lambda_dd=0.05`, `tc_bps=10`, `train_steps=50000`

En esta corrida **solo se varía la función de recompensa** (`log_return`, `turnover_penalized`, `drawdown_penalized`, `combined`). Cada una se entrena con 5 semillas × 5 ventanas walk-forward = **25 entrenamientos por recompensa**.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

plt.rcParams['figure.figsize'] = (11, 5)
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3
pd.set_option('display.float_format', lambda x: f'{x:,.3f}')

REWARDS = ['log_return', 'turnover_penalized', 'drawdown_penalized', 'combined']
TAG = 'tc0_02_dd0_05__seeds0-1-2-3-4__rewardcmp'
PATH = '.'  # los parquet deben estar junto al notebook  # ajustar si los parquet están en otra carpeta
COLORS = {'log_return':'#d62728','turnover_penalized':'#1f77b4',
          'drawdown_penalized':'#9467bd','combined':'#2ca02c'}
```

## 1. Carga de datos

```python
# Leaderboard agregado
leaderboard = pd.read_parquet(f'{PATH}/rewardcmp_1h.parquet')

# Summaries (una fila por seed × window × estrategia)
summaries = {}
for r in REWARDS:
    summaries[r] = pd.read_parquet(f'{PATH}/summary_1h__{r}__{TAG}.parquet')
summary_all = pd.concat(summaries.values(), ignore_index=True)

print('Leaderboard shape:', leaderboard.shape)
print('Summary combinado shape:', summary_all.shape)
print('Estrategias:', sorted(summary_all.strategy.unique()))
print('Ventanas:', sorted(summary_all.window.unique()))
print('Semillas:', sorted(summary_all.seed.unique()))
leaderboard
```

## 2. El leaderboard oficial usa la MEDIA — y eso engaña

El leaderboard ordena por `mean_sortino`. El problema: los Sortino del agente entrenado varían enormemente entre semillas (de valores negativos a +100), así que **unos pocos outliers afortunados inflan el promedio**. Comparamos media vs mediana y % de ventanas rentables, mirando solo `DQN_trained` (el agente que nos interesa).

```python
dqn = summary_all[summary_all.strategy == 'DQN_trained'].copy()

robust = dqn.groupby('reward_type').agg(
    n=('sortino','size'),
    sortino_mean=('sortino','mean'),
    sortino_median=('sortino','median'),
    sortino_std=('sortino','std'),
    cumret_mean=('cum_ret','mean'),
    cumret_median=('cum_ret','median'),
    maxdd_mean=('max_dd','mean'),
    turnover_mean=('total_turnover','mean'),
    fees_mean=('approx_fees_frac','mean'),
).reindex(REWARDS)
robust['pct_ventanas_positivas'] = dqn.groupby('reward_type').apply(
    lambda g: (g.cum_ret > 0).mean()*100, include_groups=False).reindex(REWARDS)

robust
```

### Media vs mediana del Sortino — la diferencia que cambia el ranking

```python
fig, ax = plt.subplots(figsize=(11,5))
x = np.arange(len(REWARDS)); w = 0.38
ax.bar(x - w/2, robust['sortino_mean'], w, label='Media (leaderboard)', color='#bbbbbb')
ax.bar(x + w/2, robust['sortino_median'], w, label='Mediana (robusta)',
       color=[COLORS[r] for r in REWARDS])
ax.axhline(0, color='k', lw=0.8)
ax.set_xticks(x); ax.set_xticklabels(REWARDS, rotation=15)
ax.set_ylabel('Sortino'); ax.set_title('Sortino del DQN entrenado: media vs mediana')
ax.legend()
for i,r in enumerate(REWARDS):
    ax.annotate(f"{robust['sortino_median'][r]:.1f}",
                (i+w/2, robust['sortino_median'][r]),
                ha='center', va='bottom' if robust['sortino_median'][r]>=0 else 'top', fontsize=9)
plt.tight_layout(); plt.show()
```

**Lectura:** por mediana, `combined` y `turnover_penalized` quedan claramente positivas, mientras `log_return` y `drawdown_penalized` caen a Sortino negativo — lo contrario de lo que sugiere el leaderboard, donde `log_return` aparecía 2º.

## 3. Distribución del Sortino por semilla (por qué la media miente)

```python
fig, ax = plt.subplots(figsize=(11,5))
data = [dqn[dqn.reward_type==r].sortino.values for r in REWARDS]
bp = ax.boxplot(data, labels=REWARDS, showmeans=True, patch_artist=True)
for patch, r in zip(bp['boxes'], REWARDS):
    patch.set_facecolor(COLORS[r]); patch.set_alpha(0.5)
ax.axhline(0, color='k', lw=0.8)
ax.set_ylabel('Sortino'); ax.set_title('Dispersión del Sortino entre las 25 corridas (5 seeds × 5 ventanas)')
plt.xticks(rotation=15); plt.tight_layout(); plt.show()
```

```python
# Strip plot: cada punto es una corrida (seed × window)
fig, ax = plt.subplots(figsize=(11,5))
for i, r in enumerate(REWARDS):
    y = dqn[dqn.reward_type==r].sortino.values
    xj = np.random.normal(i, 0.06, size=len(y))
    ax.scatter(xj, y, alpha=0.7, color=COLORS[r], s=40)
ax.axhline(0, color='k', lw=0.8)
ax.set_xticks(range(len(REWARDS))); ax.set_xticklabels(REWARDS, rotation=15)
ax.set_ylabel('Sortino'); ax.set_title('Cada punto = una corrida. La cola alta arrastra la media')
plt.tight_layout(); plt.show()
```

## 4. ¿Opera con sensatez? Turnover y comisiones

```python
fig, axes = plt.subplots(1, 2, figsize=(13,5))
axes[0].bar(REWARDS, robust['turnover_mean'], color=[COLORS[r] for r in REWARDS])
axes[0].axhline(1800, color='red', ls='--', lw=1, label='~SMA crossover (1.800)')
axes[0].set_title('Turnover total medio'); axes[0].set_ylabel('turnover')
axes[0].legend(); axes[0].tick_params(axis='x', rotation=15)

axes[1].bar(REWARDS, robust['fees_mean'], color=[COLORS[r] for r in REWARDS])
axes[1].set_title('Comisiones aprox. (fracción del capital)'); axes[1].set_ylabel('fees_frac')
axes[1].tick_params(axis='x', rotation=15)
plt.tight_layout(); plt.show()
```

**Lectura:** `log_return` y `drawdown_penalized` sobre-operan (turnover ~11.000, comisiones ~11× el capital), lo que destruye el retorno neto. Las penalizaciones de turnover (`turnover_penalized` y `combined`) mantienen la operativa en niveles realistas (~1.400–1.600).

## 5. % de ventanas rentables y comparación con baselines

```python
fig, ax = plt.subplots(figsize=(11,5))
ax.bar(REWARDS, robust['pct_ventanas_positivas'], color=[COLORS[r] for r in REWARDS])
ax.axhline(50, color='k', ls=':', lw=1)
ax.set_ylabel('% ventanas con retorno > 0')
ax.set_title('Consistencia: % de las 25 corridas con retorno acumulado positivo')
ax.tick_params(axis='x', rotation=15)
for i,r in enumerate(REWARDS):
    ax.annotate(f"{robust['pct_ventanas_positivas'][r]:.0f}%", (i, robust['pct_ventanas_positivas'][r]),
                ha='center', va='bottom')
plt.tight_layout(); plt.show()
```

```python
# DQN_trained vs baselines, emparejado por (seed, window)
def beats(df, base):
    piv = df.pivot_table(index=['seed','window'], columns='strategy', values='cum_ret')
    return (piv['DQN_trained'] > piv[base]).mean()*100

rows = []
for r in REWARDS:
    df = summaries[r]
    rows.append({'reward': r,
                 'vs Equal weight': beats(df,'Equal weight'),
                 'vs Hold Asset_0': beats(df,'Hold Asset_0'),
                 'vs SMA crossover': beats(df,'SMA crossover')})
beats_df = pd.DataFrame(rows).set_index('reward')
print('% de ventanas donde el DQN entrenado SUPERA a cada baseline:')
beats_df
```

```python
fig, ax = plt.subplots(figsize=(11,5))
x = np.arange(len(REWARDS)); w = 0.25
for k,(col,c) in enumerate(zip(beats_df.columns, ['#2ca02c','#ff7f0e','#d62728'])):
    ax.bar(x + (k-1)*w, beats_df[col], w, label=col, color=c)
ax.axhline(50, color='k', ls=':', lw=1)
ax.set_xticks(x); ax.set_xticklabels(REWARDS, rotation=15)
ax.set_ylabel('% ventanas ganadas'); ax.set_title('DQN entrenado vs baselines (emparejado por seed+ventana)')
ax.legend(); plt.tight_layout(); plt.show()
```

**Lectura:** `combined` gana al equal-weight en el 64% de las ventanas y `turnover_penalized` en 48%. Pero **ninguna recompensa supera al SMA crossover en ninguna ventana (0%)**: el SMA es un benchmark muy fuerte en estos datos. El `beats_eqw=1.0` del leaderboard se refiere al equal-weight, no al SMA.

## 6. Curvas de equity de la mejor corrida por recompensa

```python
# Para cada recompensa, tomamos la mejor combinación (seed, window) por cum_ret final
# y graficamos la curva de portfolio_value del DQN_trained.
fig, axes = plt.subplots(2, 2, figsize=(14, 9))
for ax, r in zip(axes.flat, REWARDS):
    steps = pd.read_parquet(f'{PATH}/steps_1h__{r}__{TAG}.parquet')
    d = steps[steps.strategy=='DQN_trained']
    # mejor corrida segun valor final
    best = (d.groupby(['seed','window']).portfolio_value.last().idxmax())
    seed_b, win_b = best
    g = d[(d.seed==seed_b) & (d.window==win_b)].sort_values('t')
    ax.plot(g.t, g.portfolio_value, color=COLORS[r], lw=1.2, label='DQN_trained')
    # baseline equal weight en misma ventana/seed
    eq = steps[(steps.strategy=='Equal weight') & (steps.seed==seed_b) & (steps.window==win_b)].sort_values('t')
    ax.plot(eq.t, eq.portfolio_value, color='gray', lw=1, ls='--', label='Equal weight')
    ax.set_title(f'{r}\nmejor corrida: seed {seed_b}, {win_b}', fontsize=10)
    ax.set_xlabel('paso'); ax.set_ylabel('valor portafolio'); ax.legend(fontsize=8)
plt.tight_layout(); plt.show()
```

```python
# Distribución del valor final del portafolio (todas las corridas) — escala log
fig, ax = plt.subplots(figsize=(11,5))
finals = {}
for r in REWARDS:
    finals[r] = (dqn[dqn.reward_type==r].cum_ret + 1).clip(lower=1e-3)
parts = ax.violinplot([np.log10(finals[r]) for r in REWARDS], showmedians=True)
ax.set_xticks(range(1,len(REWARDS)+1)); ax.set_xticklabels(REWARDS, rotation=15)
ax.axhline(0, color='k', lw=0.8, label='break-even (capital intacto)')
ax.set_ylabel('log10(valor final / capital inicial)')
ax.set_title('Distribución del resultado final por recompensa (log)')
ax.legend(); plt.tight_layout(); plt.show()
```

## 7. Conclusiones

**Sobre la configuración probada.** Esta corrida usa una única combinación de hiperparámetros (`lambda_tc=0.02`, `lambda_dd=0.05`, `tc_bps=10`, `train_steps=50000`). Lo que se compara aquí son las cuatro funciones de recompensa, no distintos lambdas. La búsqueda de lambdas es el stage `gridsearch`.

**El ranking del leaderboard engaña.** El leaderboard ordena por Sortino medio (`combined` > `log_return` > `turnover_penalized` > `drawdown_penalized`), pero la media está dominada por unos pocos outliers afortunados. Al usar la mediana y el % de ventanas rentables, el cuadro real es:

- **`combined` — la mejor.** Mediana de Sortino positiva (~+4.2), 80% de ventanas rentables, drawdown contenido (~-0.41), turnover bajo (~1.600) y gana al equal-weight en 64% de las ventanas.
- **`turnover_penalized` — sólida y la más barata de operar.** Mediana positiva (~+3.4), 80% de ventanas rentables, drawdown ~-0.41 y el menor turnover (~1.400).
- **`log_return` — mala pese a ir 2ª en el leaderboard.** Mediana de Sortino negativa, solo 12% de ventanas rentables, sobre-opera (turnover ~11.000, comisiones que se comen el capital).
- **`drawdown_penalized` — la peor.** Mediana negativa, 16% de ventanas rentables y, paradójicamente, no controla el drawdown (~-0.90). La penalización no funcionó a este número de pasos.

**Advertencias importantes:**
1. **Ninguna recompensa supera al SMA crossover** (0% de las ventanas en las cuatro). El RL aún queda por debajo de una regla técnica simple en estos datos.
2. **Alta varianza entre semillas** (desviaciones de Sortino de 20–80) con solo 5 semillas y entrenamiento corto (50K pasos): los números son ruidosos y no deben tomarse como definitivos.
3. A veces el `DQN_untrained` ya da resultados positivos, lo que sugiere que parte del "éxito" puede venir del azar en inicialización/datos más que del aprendizaje.

**Recomendación.** Para el gridsearch de lambdas, descartar `log_return` y `drawdown_penalized` y continuar con **`combined`** (la más prometedora) y `turnover_penalized` como respaldo. Antes de sacar conclusiones firmes, subir `train_steps` y revisar por qué ninguna recompensa se acerca al SMA.
